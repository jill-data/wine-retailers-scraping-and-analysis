```python
stop = False
posting_urls = []
while not stop:
    postings = driver.find_elements(By.XPATH,'//a[contains(@class,"h3")]')
    print("Found " + str(len(postings)) + " postings")
    for posting in postings:
        posting_urls.append(posting.get_attribute('href'))
    next_arrows = driver.find_elements(By.XPATH,'//a[contains(@id,"nextPage")]')
    if len(next_arrows) > 0:
        next_arrow = next_arrows[0]
        if next_arrow.is_enabled():
            next_arrow.click()
            time.sleep(1) # We need to wait a little, as the page takes time to load (try setting the wait to 0 and see what happens!)
        else:
            stop = True
    else:
        stop = True
```

# The main step

```python
ABV =[]
Size = []
Types = []
Names = []
years = []
prices = []
Country = []
Ratings = []
reviews = []
for i in range(len(posting_urls)):
    link = posting_urls[i]
    page = requests.get(link)
    soup = BeautifulSoup(page.content, "html.parser")
    
    # Countries of the wines
    cy = soup.find("a", class_="emp")
    if not cy:
        cy = 'Not found'
        Country.append(cy)
    else:
        cy = soup.find("a", class_="emp").text
        cy = re.sub('\s+', ' ', cy)
        Country.append(cy)
        
        # Types of the wines
        Type = soup.find("div", class_="detail-text")
        if not Type:
            Type = 'Not found'
            Types.append(Type)
        else:
            Type = soup.find("div", class_="detail-text").text
            Type = re.sub('\s+', ' ', Type)
            Type = Type.split('-')[0]
            Types.append(Type)
        
        # Rating of the wines
        rate = soup.find("span", class_="rating-score")
        if not rate:
            rate = 'Not found'
            Ratings.append(rate)
        else:
            rate = soup.find("span", class_="rating-score").text
            rate = re.sub('\(.*?\)', '' ,rate)
            rate = re.sub('\s+', '', rate)
            rate = rate.split('/')[0]
            Ratings.append(rate)
        
        # Number of reviews
        review = soup.find("span", class_="no-reviews")
        if not review:
            review = 'Not found'
            reviews.append(review)
        else:
            review = soup.find("span", class_="no-reviews").text
            review = review.split('(')[1].split(' ')[0]
            reviews.append(review)
        
        # Years of the wines
        year = soup.find("h1", class_="prod-name")
        if not year:
            year = 'Not found'
            years.append(year)
        else:
            year = soup.find("h1", class_="prod-name").text
            year = year.split('\n')[1].split(' ')[-1]
            years.append(year)
        
        # Prices of the wines
        price = soup.find("span", class_="price-per-bottle")
        if not price:
            price = 'Not found'
            prices.append(price)
        else:
            price = soup.find("span", class_="price-per-bottle").text
            price = re.sub('\s+', '', price)
            price = float(price.replace('', ''))
            prices.append(price)
        
        # Names of the products
        Name = soup.find("h1", class_="prod-name")
        if not Name:
            Name = 'Not found'
            Names.append(Name)
        else:
            Name = soup.find("h1", class_="prod-name").text
            Name = re.sub('\s+', ' ', Name)
            Names.append(Name)
        
        # ABV of the wines     # done
        abv = soup.find("div", class_="col-lg-6 col-sm-6 col-md-6 no-pad")
        if not abv:
            abv = 'Not found'
            ABV.append(abv)
        else:
            abv = soup.find("div", class_="col-lg-6 col-sm-6 col-md-6 no-pad").find_all('li')[0].text
            abv = re.sub('\s+', ' ', abv)
            abv = (abv.replace('ABV', '').replace('%', ''))
            ABV.append(abv)
        
        # Size of the wines # Done
        size = soup.find("div", class_="col-lg-6 col-sm-6 col-md-6 no-pad")
        if not size:
            size = 'Not found'
            Size.append(size)
        else:
            size = soup.find("div", class_="col-lg-6 col-sm-6 col-md-6 no-pad").find_all('li')[2].text
            size = re.sub('\s+', ' ', size)
            size = (size.replace('ml', ''))
            Size.append(size)
        
        print(i)
        
        
wines_details = pd.DataFrame(
        {'Name': Names,
         'Country': Country,
         'Type': Types,
         'Year': years,
         'Price': prices,
         'Rating': Ratings,
         'Review': reviews,
         'Size': Size,
         'ABV': ABV})
wines_details
#print(Country, Types, Ratings, reviews, years, prices, Names)
```
